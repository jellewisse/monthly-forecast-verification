```python
import shapefile
from matplotlib.path import Path

# Load the shapefile
shapefile_path = "path_to_shapefile.shp"

# Read the shapefile
sf = shapefile.Reader(shapefile_path)

# Extract the polygon(s) from the shapefile
shapes = sf.shapes()

# Open the xarray dataset and extract relevant dimensions
data = xr.open_dataset("path_to_dataset.nc")
lat = data["lat"].values
lon = data["lon"].values

# Create an empty mask
mask = np.zeros((len(lat), len(lon)), dtype=bool)

# Iterate over each shape in the shapefile and create a mask
for geom in shapes:
    # Extract the polygon vertices
    vertices = geom.points

    # Create a matplotlib path object
    path = Path(vertices)

    # Generate a grid of points
    points = np.column_stack((lon, lat))

    # Check if each point is within the polygon
    mask |= path.contains_points(points).reshape(mask.shape)

# Apply the mask to the xarray dataset
masked_data = data.where(mask)


```
